```python
df=pd.read_excel('canada.xlsx',skiprows=20,skipfooter=2,sheet_name="Canada by Citizenship")
df.head()
```

```python
cols_to_drops=['Type','Coverage','AREA','REG','DEV']
df.drop(columns=cols_to_drops,inplace=True)
```

```python
print(df.columns.to_list())
```

```python
df.rename(columns={
    'OdName':'Country','AreaName':'Continent',
    'RegName':'Region','DevName':'Status'
},
inplace=True)
```

```python
df.head(3)
```

```python
years=list(range(1980,2014))
print(years)
```

```python
df['Total']=df[years].sum(axis=1)
```

```python
df.head(3)
```

```python
df[1980].plot(kind='bar',figsize=(10,6))
```

```python
df.set_index('Country',inplace=True)
```

```python
df.sort_values(by=1980,ascending=False).head(25)[1980].plot(kind='bar',figsize=(15,5))
```
