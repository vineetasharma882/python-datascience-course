```python
a=np.random.randint(1,10,(3,3))
b=np.random.randint(1,10,(3,3))
print(a)
print(b)
print(a+b)
print(a.max())
print(a.mean())
print(a.min())
print(a.std())
print(a.sum())
print(a.max(axis=0))
print(a.mean(axis=0))
print(a.min(axis=0))
print(a.std(axis=0))
print(a.sum(axis=0))
print(a.max(axis=1))
print(a.mean(axis=1))
print(a.min(axis=1))
print(a.std(axis=1))
print(a.sum(axis=1))
```

```python
t=np.arange(-np.pi*3,np.pi,0.05)
x=np.sin(t)
y=np.cos(t)
print(x)
print(y)
print(np.pi)
```

```python
import matplotlib.pyplot as plt
plt.plot(t,x,'r--o')
plt.plot(t,y,'g:x')
plt.title("vsualization")
plt.show()
```
