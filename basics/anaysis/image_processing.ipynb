```python
import requests
from PIL import Image
from io import BytesIO
```

```python
def download_img(url):
    response= requests.get(url)
    img=Image.open(BytesIO(response.content))
    return img
```

```python
im1=download_img('https://cdn.pixabay.com/photo/2019/09/12/13/29/sunset-4471603_1280.jpg')
im1.save('image.png')
```
