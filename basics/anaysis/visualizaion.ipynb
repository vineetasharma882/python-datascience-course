```python
# random line plot
plt.figure(figsize=(15,6), dpi=300) #dpi=dots per inch
x= np.linspace(0,10,100)
y= np.random.randn(100)
_= plt.plot(x,y, linewidth=2, color='red', linestyle='--',marker='o',markersize=10)
```

```python
plt.figure(figsize=(15,4),dpi=300)
x=np.random.randint(0,10,25)
y=np.random.randint(0,10,25)
plt.grid(False)
_=plt.bar(x,y,color='red')
```
